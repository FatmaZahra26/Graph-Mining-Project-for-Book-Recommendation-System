##                  *GNN : Graph Neural Network*
Un réseau de neurones graphique (GNN) est un type de modèle d'apprentissage automatique qui peut être utilisé pour traiter des données structurées sous forme de graphes. Les graphes sont des structures de données qui représentent des relations entre des objets ou des concepts, où les objets sont représentés par des nœuds et les relations par des arêtes. Les GNN sont conçus pour traiter ces données en utilisant des mécanismes d'agrégation de voisinage pour combiner les informations des nœuds voisins d'un nœud donné. Les GNN peuvent être utilisés pour des tâches telles que la classification de nœuds, la prédiction de lien ou la génération de graphes. Les GNNs sont particulièrement utiles pour les tâches de traitement de données liées aux graphes comme les systèmes de recommandations, la reconnaissance de la communauté, l'analyse de la propagation de l'information dans les réseaux sociaux.

##                  *Un Système de recommandation*
Un système de recommandation est un système qui cherche à prédire ou à filtrer les préférences en fonction des choix de l'utilisateur. Les systèmes de recommandation sont utilisés dans une variété de domaines, tels que les films, la musique, les nouvelles, les livres, les articles de recherche, les requêtes de recherche, les tags sociaux et les produits en général. Ils permettent de développer les idées en fournissant des suggestions personnalisées en fonction des préférences de l'utilisateur. Les systèmes de recommandation basés sur les réseaux de neurones graphiques (GNN) permettent de modéliser les relations entre les utilisateurs et les éléments (comme les livres) qu'ils ont consommé, et d'utiliser ces relations pour prédire les éléments que l'utilisateur aimera dans l'avenir.

## *LightGCN*

* **LightGCN** est une version allégée des réseaux de convolution sur les graphes (GCN), un type de modèle d'apprentissage profond pour les données structurées en graphe. LightGCN est conçu pour être plus efficace et plus rapide que les modèles GCN traditionnels, tout en conservant des performances similaires ou meilleures.

* **L'idée principale derrière LightGCN** est de simplifier l'architecture GCN et de réduire le nombre de paramètres en remplaçant la multiplication de matrice de poids pleine par une multiplication de matrice creuse. Cela permet au modèle d'apprendre une représentation compacte et plus efficace de la structure du graphe.

* Le bloc de construction de base de LightGCN est la couche LightGCN, qui est une version simplifiée de la couche GCN. Chaque couche LightGCN possède une matrice de poids creuse qui est utilisée pour calculer les caractéristiques de sortie de la couche. La matrice de poids creuse est obtenue en utilisant une étape de prétraitement qui sélectionne un sous-ensemble des arêtes les plus importantes dans le graphe.Il utilise également une stratégie de normalisation différente appelée **normalisation de graphe par arête** , qui au lieu d'utiliser la matrice de degré comme facteur de normalisation, utilise la matrice identité plus la matrice d'adjacence normalisée par arête.

>  Le modèle LightGCN peut être formé sur un grand graphe en empilant plusieurs couches LightGCN ensemble et en les formant de bout en bout. LightGCN s'est avéré efficace dans une variété de tâches basées sur les graphes, telles que la classification de nœuds, la prédiction de lien et la classification de graphes.

* En outre, étant donné que l'objectif principal des schémas de filtrage collaboratif est de trouver d'autres utilisateurs similaires sur la base de leurs préférences, la simple diffusion des intégrations de nœuds le long du graphe via l'agrégation de voisinage est suffisante pour capturer les informations nécessaires à la création d'intégrations appropriées pour prédire si un utilisateur aimera un article. Le diagramme suivant montre une façon simple de visualiser cela, avec les utilisateurs à gauche et les articles à droite, et nous pouvons voir que l'intégration du nœud en surbrillance repose sur des nœuds situés à K sauts, selon le nombre de couches utilisées dans le modèle LightGCN.

Voici l'architecture du Light GCN :



Entre chaque couche, le LightGCN utilise la règle de propagation suivante pour l'intégration des utilisateurs et des éléments.

\begin{equation}
e_u^{(k+1)} = \sum_{i \in N_u} \frac{1}{\sqrt{|N_u|}\sqrt{|N_i|}} e_i^{(k)} \quad e_i^{(k+1)} = \sum_{u \in N_i} \frac{1}{\sqrt{|N_i|}\sqrt{|N_u|}} e_u^{(k)}
\end{equation}


$N_u$ : l'ensemble de tous les voisins de l'utilisateur $u$ (éléments aimés par $u$)

$N_i$ : l'ensemble de tous les voisins de l'élément $i$ (utilisateurs qui ont aimé $i$)

$e_u^{(k)}$ : intégration de l'utilisateur de la k-ième couche

$e_i^{(k)}$ : intégration de l'élément de la k-ième couche


## Combinaison de couches et prédiction du modèle
Les seuls paramètres entraînables du LightGCN sont les embeddings de la couche 0, $e_u^{(0)}$ et $e_i^{(0)}$ pour chaque utilisateur et chaque élément. Nous combinons les incorporations obtenues à chaque couche de propagation pour former les incorporations finales pour tous les utilisateurs et éléments, $e_u$ et $e_i$ via l'équation suivante.


\begin{equation}
e_u = \sum_{k = 0}^K \alpha_k e_u^{(k)} \quad e_i = \sum_{k = 0}^K \alpha_k e_i^{(k)}
\end{equation}

$\alpha_k$ : hyperparamètre qui pondère la contribution de l'incorporation de la k-ième couche à l'incorporation finale.

La prédiction du modèle est obtenue en prenant le produit interne des encastrements finaux de l'utilisateur et de l'élément.

\begin{equation}
\hat{y}_{ui} = e_u^Te_i
\end{equation}

## Forme matricielle
Dans notre implémentation, nous utilisons la forme matricielle du LightGCN. Nous effectuons une diffusion multi-échelle pour obtenir l'incorporation finale, qui additionne les incorporations diffusées à travers les échelles multi-sauts. 

\begin{equation}
E^{(K)} = \alpha_0 E^{(0)} + \alpha_1 \tilde{A}^1 E^{(0)} + \alpha_2 \tilde{A}^2 E^{(0)} + \cdot \cdot \cdot + \alpha_K \tilde{A}^K \tilde{A} E^{(0)}
\end{equation}

$E^{(0)} \in \mathcal{R}^{(M + N)} \times T$ : Embeddings initiaux empilés d'éléments et d'utilisateurs où $M$, $N$ et $T$ désignent respectivement le nombre d'utilisateurs, le nombre d'éléments et la dimension de chaque embedding

$\tilde{A} = D^{-\frac{1}{2}}AD^{-\frac{1}{2}}$ :Matrice d'adjacence symétriquement normalisée

Nous évaluons notre modèle à l'aide des métriques suivantes

\begin{equation}
\text{Recall} = \frac{TP}{TP + FP}

```python
class PlainData(Data):
    def __inc__(self, key, value, *args, **kwargs):
        return 0

class BookDataSet(Dataset):
  def __init__(self, edge_index, transform = None, pre_transform = None, pre_filter = None):
    self.edge_index = edge_index
    self.user_name = torch.unique(edge_index[0, :]).tolist()
    self.num_nodes = len(self.user_name)
    super().__init__(transform, pre_transform, pre_filter)

  def len(self):
    return self.num_nodes

  def get(self, idx):
    edge_index = self.edge_index[:, self.edge_index[0, :] == idx] 
    return PlainData(edge_index = edge_index) 
```

```python
def sample_negative_edges(batch, num_users, num_nodes):
    negs = torch.randint(num_users, num_nodes, (batch.edge_index.shape[1],)) # Sample negative edges
    edge_index_negs = torch.stack([batch.edge_index[0,:], negs], dim=0) # Stack indices for negative edges
    return Data(edge_index=edge_index_negs)
```

Cette fonction, sample_negative_edges, est utilisée pour échantillonner les bords négatifs de l'ensemble de données d'apprentissage. Elle prend les entrées suivantes :

* batch : un objet PyTorch Data qui contient les arêtes positives de l'ensemble de données d'entraînement.

* num_users : le nombre d'utilisateurs dans l'ensemble de données.

* num_nodes : le nombre de noeuds dans le graphe.

La fonction crée une liste vide negs qui stockera les bords négatifs. Ensuite, pour chaque arête positive dans le lot, elle génère un nombre aléatoire entre num_users et num_nodes en utilisant la fonction torch.randint, qui représentera l'indice d'une arête négative. Elle ajoute ensuite le nombre aléatoire à la liste negs.

Enfin, la fonction crée un nouvel objet Data PyTorch, edge_index_negs, qui contient les indices des arêtes négatives en empilant les indices des arêtes positives et des arêtes négatives stockées dans la liste negs. Elle renvoie le nouvel objet Data contenant les arêtes négatives.

Il est important de noter que cette fonction échantillonne les arêtes négatives de façon aléatoire et qu'elle n'est peut-être pas le meilleur choix pour tous les cas d'utilisation, surtout si vous avez une connaissance préalable des arêtes négatives.

Cette fonction prend un lot d'arêtes, représenté par un objet Data, ainsi que le nombre total d'utilisateurs et le nombre total de noeuds dans le graphe. Elle échantillonne des arêtes négatives pour chaque arête positive du lot en sélectionnant aléatoirement un noeud pour le second point d'extrémité de chaque arête, à condition que ce noeud ne soit pas déjà un voisin du premier point d'extrémité de l'arête. Elle renvoie un nouvel objet Data avec les mêmes indices d'arêtes que le lot d'entrée, mais avec les échantillons négatifs comme second point d'extrémité.


La fonction train est un exemple de fonction d'entraînement pour un système de recommandation utilisant LightGCN. La fonction prend plusieurs arguments en entrée :

* model : le modèle LightGCN qui sera entraîné.

* train_loader : un objet PyTorch DataLoader qui charge les données d'entraînement par lots.

* train_edge : un tenseur clairsemé PyTorch qui stocke les données d'entraînement dans un format clairsemé.

* validation_loader : un objet PyTorch DataLoader qui charge les données de validation par lots.

* validation_edge : un tenseur clairsemé PyTorch qui stocke les données de 
* validation dans un format clairsemé.

* num_nodes : le nombre de noeuds dans le graphe.

* num_users : le nombre de listes de lecture dans l'ensemble de données.

* optimizer : l'optimiseur qui sera utilisé pour mettre à jour les paramètres du modèle pendant la formation.

device : le dispositif (CPU ou GPU) sur lequel le modèle sera entraîné.

* k : le nombre de recommandations supérieures qui seront faites pour chaque liste de lecture.

* n_epochs : le nombre de fois que l'ensemble des données d'entraînement sera passé dans le modèle.

> La fonction entraîne ensuite le modèle pendant le nombre d'époques n_epochs et échantillonne les bords négatifs de l'ensemble de données d'entraînement à l'aide de la fonction sample_negative_edges. Elle calcule ensuite la perte moyenne de l'ensemble de données d'entraînement, et le taux de rappel de validation, qui est le taux de rappel moyen de l'ensemble de données de validation. Enfin, il trace un graphique de la perte de formation et de la perte de validation et renvoie la perte de validation moyenne et le taux de rappel.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_fscore_support

def train(model, train_loader, train_edge, validation_loader,validation_edge, num_nodes, num_users, optimizer,  k, n_epochs,device):
    train_losses = []
    validation_losses = []
    running_loss = 0.
    total_edges = 0
    train_edge = train_edge.to(device)
    validation_edge =validation_edge.to(device)
    recalls = {}
  

    for epoch in range(1, n_epochs + 1):

        model.train()  # Training
        total_edges = 0
        for i, batch in enumerate(train_loader):
            optimizer.zero_grad()
            negative_values = sample_negative_edges(batch, num_users, num_nodes)
            batch, negative_values = batch.to(device), negative_values.to(device)
            positive_rank = model(train_edge.edge_index, batch.edge_index)  
            negative_rank = model(train_edge.edge_index, negative_values.edge_index) 
            
            loss = model.recommendation_loss(positive_rank, negative_rank)#Calcul Loss

            loss.backward()
            optimizer.step()

            running_loss += loss.item() * positive_rank.size(0) * batch.edge_index.shape[1]
            total_edges += batch.edge_index.shape[1]
            avg_loss = running_loss / total_edges
            train_losses.append(avg_loss)
            print(f'Epoch [{epoch}/{n_epochs}] - Training loss = {avg_loss}')
            model.eval() 
            with torch.no_grad():
                for j, batch in enumerate(validation_loader):
                    batch = batch.to(device)
                    TopK = model.recommend(validation_edge.edge_index, torch.unique(batch.edge_index[0, :]), torch.unique(batch.edge_index[1, :]), k)
                    # Calculating recall@k
                    unique_playlists = torch.unique(batch.edge_index[0, :])
                    for i, pidx in enumerate(unique_playlists):
                        ground_truth =validation_edge.edge_index[1,validation_edge.edge_index[0, :] == pidx].cpu()
                        preds = TopK[i].cpu()
                        if len(ground_truth) == 0:
                            recall = 0
                        else:
                            recall = len(np.intersect1d(ground_truth, preds)) / len(ground_truth)
                        recalls[pidx] = recall
                avg_val_loss = np.mean(list(recalls.values()))
                print(f'Epoch [{epoch}/{n_epochs}] - Validation recall = {avg_val_loss}')

    return print("avg_val_loss: ",avg_val_loss,"recall :",recall)
```

```python
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
from torch_geometric.nn.models import LightGCN
model = LightGCN(num_nodes = data.num_nodes, embedding_dim = 64, num_layers = 4)
model.to(device)
optimizer = torch.optim.Adam(model.parameters(), lr = 0.01)
train(model, train_loader, train_edge, validation_loader, validation_edge, data.num_nodes,num_users ,optimizer, 150, 8, device)
```
